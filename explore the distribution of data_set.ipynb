## Frequency counts of class of plants

```python
df_class=df['class']
```

```python
# Count the frequency of each plant family
class_counts = df_class.value_counts()
```

```python
# Set the threshold for the minimum frequency
threshold = 15
# Select the plant families with a frequency above the threshold
selected_class = class_counts[class_counts > threshold]

# Calculate the total frequency of the remaining plant families
other_class = class_counts[class_counts <= threshold].sum()

# Add the "other" category to the selected plant families
selected_class = selected_class.append(pd.Series([other_class], index=['other']))
```

## visualisation of family variable.

```python
df_family=df['family']
#count the frequency of the labels in family variable
family_counts = df_family.value_counts()
# Set the threshold for the minimum frequency
threshold = 50
# Select the plant families with a frequency above the threshold
selected_families = family_counts[family_counts > threshold]

# Calculate the total frequency of the remaining plant families
other_families = family_counts[family_counts <= threshold].sum()

# Add the "other" category to the selected plant families
selected_families = selected_families.append(pd.Series([other_families], index=['other']))
```

## pie chart

```python
# Set the labels to the selected plant families
labels = selected_families.index

# Set the values to the frequency of each selected plant family
values = selected_families.values

# Create a pie chart using the labels and values
plt.pie(values, labels=labels)

# Set the chart title
plt.title('Plant family Distribution')

# Show the plot
plt.show()
```
