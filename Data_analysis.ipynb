#### Considering class as dependent variable and other as independent variable(for now skip scitific name)

```python
# Create a contingency table
table = pd.crosstab(data['species'], [data['order'], data['genus'],data['family'],data['class'],data['scientificName']])
```

```python
# Perform the chi-squared test
stat, p, dof, expected = chi2_contingency(table)
```

```python
# Print the p-value
print("p-value: ", p)
```

### Using dython to find correlation by correaltion matrix and heat map then

```python
from dython.nominal import associations
```

```python
df=pd.read_csv("C:/Users/rakes/Desktop/sample.csv")
```

```python
complete_correlation= associations(df, filename= 'complete_correlation.png', figsize=(20,20))
```

```python
df_complete_corr=complete_correlation['corr']
df_complete_corr.dropna(axis=1, how='all').dropna(axis=0, how='all').style.background_gradient(cmap='coolwarm', axis=None).set_precision(2)
```
