## Now making a classification model on the data_set

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,f1_score,recall_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
```

### 1.first converting categorical string values to numeric so that classification algo can work on it

```python
#using one_hot encoding to convert categorical variable to numeric
ohe=OneHotEncoder()
```

```python
X=ohe.fit_transform(df[['genus','family','order','species','class','scientificName']])
```

### 2. Now, to apply classification algorithm (random forest algorithm) using test and train set

### considering scientificName as dependent variable and other as independent variable.

```python
# Split the data into independent variables and dependent variable
y = X[:, -len(df['scientificName'].unique())] # separate the last columns as the dependent variable
X = X[:, :-len(df['scientificName'].unique())] # rest of the columns as independent variables
y=y.toarray()#to convert he sparse matrix to dense matrix
```

```python
## Split the data into training and testing sets
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
```

```python
## fitting the random forest classifier to the training data
clf=RandomForestClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
y_pred
```

```python
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy: ",accuracy*100)
```

```python
f1=f1_score(y_test,y_pred,average='micro')
print('f1 score: ',f1*100)
```

```python
# Calculate the recall
recall = recall_score(y_test, y_pred,average='micro')
print('Recall: ', recall*100)
```

## Now using decision tree classification algorithm 

### Here considering class as dependent variable and other as independent varaible.

```python
y=ohe.fit_transform(df[['class']])
X=ohe.fit_transform(df[['genus','family','species','order']])
X=X.toarray()
y=y.toarray()
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Create a decision tree classifier
clf = DecisionTreeClassifier()
```

```python
# Train the classifier on the training data
clf.fit(X_train, y_train)
```

```python
# Make predictions on the testing data
y_pred = clf.predict(X_test)
```

```python
# Evaluate the accuracy of the classifier
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc*100)
```

```python
#f1 score combines precision and recall accuracy of the model
f1=f1_score(y_test,y_pred,average='micro')
print('f1 score: ',f1*100)
```
