```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

from sklearn.linear_model import LogisticRegression
#these below libraries are use for decision tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import f1_score,recall_score
```

```python
df=pd.read_csv("C:/Users/rakes/Desktop/sample.csv")
```

## Predicting or classifying a plant Scientific Name (Dependent variable) on the basis of attribute(independent variable)

### 1.first converting categorical string values to numeric so that classification algo can work on it

```python
#using one hot encoding to convert categorical variable to numeric
ohe=OneHotEncoder()
```

```python
X=ohe.fit_transform(df[['genus','family','order','species','class','scientificName']])
```

### 2. Now, to apply classification algorithm (random forest algorithm) using test and train set

```python
# Split the data into independent variables and dependent variable
y = X[:, -len(df['scientificName'].unique())] # separate the last columns as the dependent variable
X = X[:, :-len(df['scientificName'].unique())] # rest of the columns as independent variables
y=y.toarray()
```

```python
## Split the data into training and testing sets
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
```

```python
## fitting the random forest classifier to the training data
clf=RandomForestClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
y_pred
```

```python
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy: ",accuracy)
```

```python
f1=f1_score(y_test,y_pred,average='micro')
print('f1 score: ',f1)
```

```python
# Calculate the recall
recall = recall_score(y_test, y_pred,average='micro')
print('Recall: ', recall)
```

##  Predicting class(as dependent variable ) and other as independent variable using decision tree

```python
## this time the sample has 100 observation
df=pd.read_csv("C:/Users/rakes/Desktop/sample.csv")
```

```python
df
```

```python
#using one hot encoding to convert categorical variable to numeric
ohe=OneHotEncoder()
```

```python
X=ohe.fit_transform(df[['genus','family','order','species','class']])
```

```python
# Split the data into independent variables and dependent variable
y = X[:, -len(df['class'].unique())] # separate the last columns as the dependent variable
X = X[:, :-len(df['class'].unique())] # rest of the columns as independent variables
X=X.toarray()
y=y.toarray()
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Create a decision tree classifier
clf = DecisionTreeClassifier()
```

```python
# Train the classifier on the training data
clf.fit(X_train, y_train)
```

```python
# Make predictions on the testing data
y_pred = clf.predict(X_test)
```

```python
# Evaluate the accuracy of the classifier
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)
```

```python
f1=f1_score(y_test,y_pred,average='micro')
print('f1 score: ',f1)
```
